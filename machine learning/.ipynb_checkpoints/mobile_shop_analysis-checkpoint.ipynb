# Mobile Shop Sales Analysis Tutorial for Beginners
This tutorial is designed for beginner Python users to learn business analytics using a mobile phone repair and accessories shop dataset.

**Dataset**: Mobile shop transaction data (2022-2025)
**Objective**: Analyze sales performance using Total_Amount as the key performance indicator (KPI)

## Step 1: Setting Up Your Environment
We'll use Python libraries to analyze the data:
- **pandas**: For handling data tables.
- **matplotlib**: For creating plots and charts.
- **seaborn**: For advanced visualizations.
- **sklearn**: For building predictive models.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from tabulate import tabulate
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

## Step 2: Loading the Data
The dataset (`mobile_shop_transactions_3years.csv`) contains transaction records with columns like Transaction_ID, Product_Name, Category, Brand, Model, Total_Amount, Order_Type, and more.

```python
# Load the CSV file
data = pd.read_csv('mobile_shop_transactions_3years.csv')

# Display the first 5 rows to understand the data
print("First 5 rows of the dataset:")
print(data.head())

# Display basic information
print("\nDataset Info:")
print(f"Total Rows: {len(data)}")
print(f"Total Columns: {len(data.columns)}")
print(f"\nColumn Names: {list(data.columns)}")
```

**Explanation**: `pd.read_csv` loads the CSV into a pandas DataFrame, like a spreadsheet. The `head()` function shows the first 5 rows to get a sense of the data structure.

## Step 3: Exploring the Data
Let's compute some basic statistics to understand the sales:
- Total sales amount
- Average transaction value
- Number of transactions
- Top-selling brand

```python
# Calculate total sales
total_sales = data['Total_Amount'].sum()

# Calculate average transaction value
avg_transaction = data['Total_Amount'].mean()

# Calculate total transactions
total_transactions = len(data)

# Find top-selling brand
top_brand = data.groupby('Brand')['Total_Amount'].sum().idxmax()

# Display results
print(f"Total Sales: RM {total_sales:,.2f}")
print(f"Average Transaction Value: RM {avg_transaction:,.2f}")
print(f"Total Transactions: {total_transactions:,}")
print(f"Top-Selling Brand: {top_brand}")
```

**Explanation**: We use `sum()` and `mean()` to calculate totals and averages. The `groupby` function groups data by Brand, and `idxmax()` finds the brand with the highest total sales.

## Step 4: Visualizing Sales by Brand
Let's create a bar plot to show total sales for each brand.

```python
# Calculate total sales by brand
brand_sales = data.groupby('Brand')['Total_Amount'].sum().sort_values(ascending=False).reset_index()

# Create a bar plot for sales by brand
plt.figure(figsize=(10, 6))
sns.barplot(x='Brand', y='Total_Amount', data=brand_sales)
plt.title('Total Sales by Brand', fontsize=16, fontweight='bold')
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Total Amount (RM)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sales_by_brand.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\nTop 3 Brands by Revenue:")
for idx, row in brand_sales.head(3).iterrows():
    print(f"{idx+1}. {row['Brand']}: RM {row['Total_Amount']:,.2f}")
```

**Explanation**: `sns.barplot` creates a bar chart. We group the data by Brand and sum Total_Amount. The plot is saved as an image and displayed.

## Step 5: Visualizing Sales by Category
Let's see how sales are distributed between Accessories and Repair services.

```python
# Calculate total sales by category
category_sales = data.groupby('Category')['Total_Amount'].sum().reset_index()

# Create a pie chart for category distribution
plt.figure(figsize=(8, 8))
plt.pie(category_sales['Total_Amount'], labels=category_sales['Category'], 
        autopct='%1.1f%%', startangle=90, colors=['#2E7D32', '#1976D2'])
plt.title('Sales Distribution by Category', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('sales_by_category.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\nCategory Breakdown:")
for idx, row in category_sales.iterrows():
    print(f"{row['Category']}: RM {row['Total_Amount']:,.2f}")
```

**Explanation**: We use a pie chart to show the percentage distribution of sales between Accessories and Repair categories.

## Step 6: Visualizing Sales by Model (Top 10)
Let's identify the top 10 phone models that generate the most revenue.

```python
# Calculate total sales by model and get top 10
model_sales = data.groupby('Model')['Total_Amount'].sum().sort_values(ascending=False).head(10).reset_index()

# Create a horizontal bar plot for top 10 models
plt.figure(figsize=(10, 8))
sns.barplot(y='Model', x='Total_Amount', data=model_sales, orient='h')
plt.title('Top 10 Models by Sales Revenue', fontsize=16, fontweight='bold')
plt.xlabel('Total Amount (RM)', fontsize=12)
plt.ylabel('Model', fontsize=12)
plt.tight_layout()
plt.savefig('sales_by_model.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\nTop 10 Models by Revenue:")
for idx, row in model_sales.iterrows():
    print(f"{idx+1}. {row['Model']}: RM {row['Total_Amount']:,.2f}")
```

**Explanation**: We group by Model, sum the Total_Amount, and use `head(10)` to get the top 10. A horizontal bar chart makes it easier to read model names.

## Step 7: Visualizing Sales by Order Type
Let's analyze how sales are distributed across different sales channels (Walk-In, Shopee, Lazada).

```python
# Calculate total sales by order type
order_sales = data.groupby('Order_Type')['Total_Amount'].sum().sort_values(ascending=False).reset_index()

# Create a bar plot for order types
plt.figure(figsize=(10, 6))
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
sns.barplot(x='Order_Type', y='Total_Amount', data=order_sales, palette=colors)
plt.title('Total Sales by Order Type', fontsize=16, fontweight='bold')
plt.xlabel('Order Type', fontsize=12)
plt.ylabel('Total Amount (RM)', fontsize=12)
plt.tight_layout()
plt.savefig('sales_by_order_type.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\nSales by Order Type:")
for idx, row in order_sales.iterrows():
    print(f"{row['Order_Type']}: RM {row['Total_Amount']:,.2f}")
```

**Explanation**: `value_counts()` counts transactions per order type. The bar plot shows which sales channel generates the most revenue.

## Step 8: Visualizing Sales Over Time
We'll group sales by month to see trends over time.

```python
# Convert Transaction_DateTime to datetime
data['Transaction_DateTime'] = pd.to_datetime(data['Transaction_DateTime'], format='%d/%m/%Y %H:%M')

# Extract year and month for grouping
data['YearMonth'] = data['Transaction_DateTime'].dt.to_period('M')

# Group by month and sum sales
monthly_sales = data.groupby('YearMonth')['Total_Amount'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()

# Create a line plot for monthly sales
plt.figure(figsize=(14, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Total_Amount'], marker='o', linewidth=2, markersize=6)
plt.title('Monthly Sales Trend (2022-2025)', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Amount (RM)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('monthly_sales_trend.png', dpi=300, bbox_inches='tight')
plt.show()
```

**Explanation**: We convert Transaction_DateTime to a datetime format, extract the year and month, and sum sales by month. The line plot shows sales trends over time.

## Step 9: Building a Linear Regression Model
We want to predict Total_Amount using Quantity, Brand, Category, and Order_Type. Since Brand, Category, and Order_Type are categorical, we'll encode them as numbers.

```python
# Prepare data for regression
# Select relevant columns
regression_data = data[['Quantity', 'Brand', 'Category', 'Unit_Price', 'Total_Amount']].copy()

# Encode categorical variables
le_brand = LabelEncoder()
le_category = LabelEncoder()
le_order = LabelEncoder()

regression_data['Brand_Encoded'] = le_brand.fit_transform(regression_data['Brand'])
regression_data['Category_Encoded'] = le_category.fit_transform(regression_data['Category'])

# Define features (X) and target (y)
X = regression_data[['Quantity', 'Brand_Encoded', 'Category_Encoded', 'Unit_Price']]
y = regression_data['Total_Amount']

# Build the model
model = LinearRegression()
model.fit(X, y)

# Display model coefficients
print("Linear Regression Model Coefficients:")
print(f"Quantity: {model.coef_[0]:.2f}")
print(f"Brand: {model.coef_[1]:.2f}")
print(f"Category: {model.coef_[2]:.2f}")
print(f"Order Type: {model.coef_[3]:.2f}")
print(f"\nIntercept: {model.intercept_:.2f}")
print(f"\nModel R Score: {model.score(X, y):.4f}")

# Predict Total_Amount and create a scatter plot
linear_y_pred = model.predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(y, linear_y_pred, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # Diagonal line
plt.title('Linear Regression: Actual vs Predicted Total Amount')
plt.xlabel('Actual Total ($)')
plt.ylabel('Predicted Total ($)')
plt.savefig('linear_regression_scatter.png')
plt.show()
```

**Explanation**: `LabelEncoder` converts categorical data (like Brand names) into numbers. `LinearRegression` finds the relationship between features and Total_Amount. The coefficients show how much each feature affects sales.

## Step 10: Making Predictions
Let's use our model to predict sales for sample scenarios.

```python
# Create and train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X, y)

# Get R score
rf_r_squared = rf_model.score(X, y)

# Print results
print("Random Forest Regression Results:")
print(f"R Score: {rf_r_squared:.4f}")
# Predict Total_Amount and create a scatter plot
predictions = rf_model.predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(y, predictions, color='green')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # Diagonal line
plt.title('Random Forest: Actual vs Predicted Total Amount')
plt.xlabel('Actual Total ($)')
plt.ylabel('Predicted Total ($)')
plt.savefig('rf_regression_scatter.png')
plt.show()
```

```python
# Create a table to compare R scores
comparison_table = [
    ["Linear Regression", f"{model.score(X, y):.4f}"],
    ["Random Forest", f"{rf_r_squared:.4f}"]
]
print("\nModel Comparison:")
print(tabulate(comparison_table, headers=["Model", "R Score"], tablefmt="pretty"))
```
